```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('../images/lena.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

FILTRO DE MEDIANA

```python
def add_median_filter(image):
    blur_image = np.zeros(image.shape, dtype=np.uint8)

    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                if i == 0 or j == 0 or i == image.shape[0] - 1 or j == image.shape[1] - 1:
                    blur_image[i, j, k] = image[i, j, k]
                else: 
                    pixels = []

                    for x in range(-1, 2):
                        for y in range(-1, 2):
                            pixels.append(image[i + x, j + y, k])
                    
                    blur_image[i, j, k] = np.median(pixels)
        
    return blur_image

```

```python
image_median_filter = add_median_filter(image)
plt.imshow(image_median_filter)
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_median_filter)
plt.title('Median Filter')
plt.axis('off')
```
